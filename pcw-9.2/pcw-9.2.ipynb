# 1. Gradients

*Linear:*
$$f(a,x) = ax$$
$$f'(a,x) = a$$

*RelU:*
$$f(x) = \left\{
  \begin{array}{lr}
    x & : x > 0\\
    0 & : x <= 0
  \end{array}
\right.
$$
$$f'(x) = \left\{
  \begin{array}{lr}
    1 & : x > 0\\
    0 & : x <= 0
  \end{array}
\right.
$$

*Sigmoid:*
$$f(x)=\frac{1}{1+e^{âˆ’x}}$$

$$f'(x)=\frac{e^{-x}}{(e^{-x}+1)^2}$$

# 2. Keras

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
from keras.models import Sequential
from keras.layers import Dense
```

```python
(moon_x,moon_y) = datasets.make_moons(n_samples=10000, noise=0.05, random_state=42)
(circle_x,circle_y) = datasets.make_circles(n_samples=10000, noise=0.025, random_state=42)
```

```python
XM_train, XM_test, ym_train, ym_test = train_test_split(moon_x, moon_y, test_size=0.33, random_state=42)
XC_train, XC_test, yc_train, yc_test = train_test_split(circle_x, circle_y, test_size=0.33, random_state=42)
```

```python
model = Sequential()
model.add(Dense(12, input_dim=2, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='relu'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(XM_train, ym_train, epochs=10, batch_size=10, verbose=0)
test_results = model.evaluate(XM_test, ym_test, verbose=0)
print(f'Test results - Loss: {test_results[0]} - Accuracy: {test_results[1]*100}%')
plot_decision_regions(XM_test, ym_test, clf=model, legend=2)
plt.show()
```

```python
model_2 = Sequential()
model_2.add(Dense(10, input_dim=2, activation='sigmoid'))
model_2.add(Dense(1, activation='sigmoid'))
```

```python
model_2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_2.fit(XM_train, ym_train, epochs=10, batch_size=10, verbose=0)
test_results_2 = model_2.evaluate(XM_test, ym_test, verbose=0)
print(f'Test results - Loss: {test_results_2[0]} - Accuracy: {test_results_2[1]*100}%')
plot_decision_regions(XM_test, ym_test, clf=model_2, legend=2)
plt.show()
```

```python
model_3 = Sequential()
model_3.add(Dense(8, input_dim=2, activation='relu'))
model_3.add(Dense(1, activation='relu'))
```

```python
model_3.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_3.fit(XM_train, ym_train, epochs=10, batch_size=10, verbose=0)
test_results_3 = model_3.evaluate(XM_test, ym_test, verbose=0)
print(f'Test results - Loss: {test_results_3[0]} - Accuracy: {test_results_3[1]*100}%')
plot_decision_regions(XM_test, ym_test, clf=model_3, legend=2)
plt.show()
```

**Moon: The first model performed the best.**
